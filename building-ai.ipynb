#### Simulated annealing

One of the simplest and most effective solutions to local optima

The method is remarkably simple. Instead of only allowing changes that improve the solution (go uphill), some changes that make it worse (go downhill) are also allowed with some probability. 

The probability of allowing a downwards transition depends on two things: how much downwards it goes and a so called 'temperature.' The higher the temperature, the higher the probability of allowing a downhill move.

The idea is to start at a high temperature so that the changes are more or less random, but to gradually decrease the temperature so that eventually, the probability of going downwards becomes vanishingly small.
