## Perceptrons as Logical Operators

### AND



```python
import pandas as pd

# For AND perceptron

# b < 0
# w1 + w2 + b > 0
# w1 + b < 0
# w2 + b < 0

# Set weight1, weight2, and bias
weight1 = 2.0
weight2 = 2.0
bias = -3.0

# Inputs and outputs
test_inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
correct_outputs = [False, False, False, True]
outputs = []

# Generate and check output
for test_input, correct_output in zip(test_inputs, correct_outputs):
    linear_combination = weight1 * test_input[0] + weight2 * test_input[1] + bias
    output = int(linear_combination >= 0)
    is_correct_string = 'Yes' if output == correct_output else 'No'
    outputs.append([test_input[0], test_input[1], linear_combination, output, is_correct_string])

# Print output
num_wrong = len([output[4] for output in outputs if output[4] == 'No'])
output_frame = pd.DataFrame(outputs, columns=['Input 1', '  Input 2', '  Linear Combination', '  Activation Output', '  Is Correct'])

print(output_frame.to_string(index=False))
```

### XOR



```python
import pandas as pd

def Linear_Combination(weight1,weight2,input1,input2,bias):
    return weight1*input1 + weight2*input2 + bias

def Activation(linear_combination):
    return int(linear_combination >= 0)    

def AND_perceptron(input1, input2):    
    weight1 = 2.0
    weight2 = 2.0
    bias = -3.0        

    return Activation(Linear_Combination(weight1,weight2,input1,input2,bias))

def OR_perceptron(input1, input2):    
    weight1 = 3.0
    weight2 = 3.0
    bias = -1.0    

    return Activation(Linear_Combination(weight1,weight2,input1,input2,bias))

def NOT_perceptron(input1, input2):    
    weight1 = 3.0
    weight2 = -7.0
    bias = 3.0

    return Activation(Linear_Combination(weight1,weight2,input1,input2,bias))
```

```python
def XOR_perceptron(input1,input2):
    
    # AND operation between input1 and input 2
    res1 = AND_perceptron(input1,input2)
    # Take the NOT on the result of AND operation
    res2 = NOT_perceptron(0,res1)
    # Take the OR operation on input1 and input2
    res3 = OR_perceptron(input1,input2)
    # Take the AND operation on above res2 and res3 to get XOR
    xor_output = AND_perceptron(res2,res3)
    
    return xor_output
    
    
    
test_inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
correct_outputs = [False, True, True, False]
outputs = []

# Generate and check output
for test_input, correct_output in zip(test_inputs, correct_outputs):
    output = XOR_perceptron(test_input[0], test_input[1])
    is_correct_string = 'Yes' if output == correct_output else 'No'
    outputs.append([test_input[0], test_input[1], output, is_correct_string])

# Print output
num_wrong = len([output[3] for output in outputs if output[3] == 'No'])
output_frame = pd.DataFrame(outputs, columns=['Input 1', 'Input 2', 'Activation Output', 'Is Correct'])

print(output_frame.to_string(index=False))
```
