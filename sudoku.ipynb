PHTSTOMEFEL RING : SET Equivalence IN SUDUKO



4 sets each of 1-9 in RED and GREEN SET



```python
l1 = [4,2,5,6,7,5,1,9,9,1,4,8,5,8,2,7,6,9,4,3,1,3,8,5,3,7,2,9,8,4,2,6,6,3,7,1]

l2 = [9,1,6,2,5,7,3,4,8,5,8,4,6,2,7,6,9,2,1,3,7,1,3,4,9,8,5,3,7,5,1,4,6,8,9,2]
```

```python
from collections import Counter

print(Counter(l1))
print(Counter(l2))
print()

if Counter(l1) == Counter(l2):
    print("l1 and l2 have the same elements with the same counts.")
else:
    print("l1 and l2 do not have the same elements with the same counts.")
```
