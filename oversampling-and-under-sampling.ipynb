```python
import numpy as np

# Generate a synthetic dataset with class imbalance
np.random.seed(0)
fraudulent_samples = np.random.randn(50, 2)  # 50 samples of fraudulent transactions
non_fraudulent_samples = np.random.randn(950, 2)  # 950 samples of non-fraudulent transactions
X = np.vstack([fraudulent_samples, non_fraudulent_samples])
y = np.hstack([np.ones(50), np.zeros(950)])  # 1 for fraudulent, 0 for non-fraudulent

# Check class distribution
unique, counts = np.unique(y, return_counts=True)
class_distribution = dict(zip(unique, counts))
print(class_distribution)
```

```python
import matplotlib.pyplot as plt

# Check class distribution
unique, counts = np.unique(y, return_counts=True)
class_distribution = dict(zip(unique, counts))

# Plot class distribution
plt.bar(class_distribution.keys(), class_distribution.values())
plt.title('Initial Class Distribution')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.xticks(list(class_distribution.keys()), ['Fraudulent', 'Non-fraudulent'])
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(sampling_strategy='auto', random_state=0)
X_resampled, y_resampled = oversampler.fit_resample(X, y)
```

```python
import matplotlib.pyplot as plt

# Check class distribution after oversampling
unique_resampled, counts_resampled = np.unique(y_resampled, return_counts=True)
class_distribution_resampled = dict(zip(unique_resampled, counts_resampled))

# Plot class distribution after oversampling
plt.bar(class_distribution_resampled.keys(), class_distribution_resampled.values())
plt.title('Class Distribution After Random Over-sampling')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.xticks(list(class_distribution_resampled.keys()), ['Fraudulent', 'Non-fraudulent'])
plt.show()
```

```python
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=0)
X_resampled, y_resampled = undersampler.fit_resample(X, y)
```

```python
# Check class distribution after under-sampling
unique_resampled, counts_resampled = np.unique(y_resampled, return_counts=True)
class_distribution_resampled = dict(zip(unique_resampled, counts_resampled))

# Plot class distribution after under-sampling
plt.bar(class_distribution_resampled.keys(), class_distribution_resampled.values())
plt.title('Class Distribution After Random Under-sampling')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.xticks(list(class_distribution_resampled.keys()), ['Fraudulent', 'Non-fraudulent'])
plt.show()
```
